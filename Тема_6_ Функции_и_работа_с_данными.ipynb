Датафрейм - это таблица

Метод apply - аналог цикла, который проходит ко всем строкам датафрейма и применяет к каждой функцию

Два режима использования:
1. Передаются значения одного столбца
2. Передается вся строка целиком

Строка в Pandas - датафрейм из одной строки

```python
import pandas as pd
df = pd.DataFrame(
    {
        'user_id': [1, 2, 3],
        'clicks': [163, 130, 97],
        'orders': [2, 4, 0]
    }
)
```

```python
# Определить, кто покупает, а кто только смотрит
def watcher(param):
    return param == 0
```

```python
# Реализация метода apply (один столбец)
df['wather'] = df['orders'].apply(watcher)
df.head()
```

```python
# Определить конверсию (вся строка)
def conversion(row):
    return row['orders'] / row['clicks']
```

```python
# axis=1 - в функцию будет передана строка
# axis=0 - будет передан столбец (по умолчанию)
df['conversion'] = df.apply(conversion, axis=1)
df.head()
```

### Group by (Группировка)
Группировка - подсчет определенной метрики для каждого уникального значения заданного столбца

Алгоритм расчета:
1. Каждому уникальному значению делаем свой датафрейм
2. Применяем функцию (например, суммы) в каждому датафрейму
3. Результаты склеиваются в один датафрейм

Пример группировки:
* Подсчет суммы продаж по странам
* Подсчет средней стоимости продаж по странам

```python
df = pd.DataFrame({'order_id': [1, 2, 3, 4, 5], 'country': ['Россия', 'Китай', 'Китай', 'Россия', 'Россия'], 
                   'category': ['Электроника', 'Авто', 'Электроника', 'Авто', 'Авто'], 
                   'amount': [100, 80, 90, 140, 90]})
df
```

```python
# Решение задачи вручную
df_russia = df[df['country'] == 'Россия']
print(df_russia)
df_china = df[df['country'] == 'Китай']
print(df_china, "\n")

# Посчитаем среднюю стомость заказа, разницу между макс и мин стоимостью заказа
print("Средняя для России", df_russia.amount.mean())
print("Размах для России", df_russia.amount.max() - df_russia.amount.min())
print("Средняя для Китая", df_china.amount.mean())
print("Размах для Китая", df_china.amount.max() - df_china.amount.min())
```

```python
# Объединим процесс разбиения на датафреймы
def groupby_func(data):
    return data.amount.max() - data.amount.min()
```

```python
df.groupby('country').apply(groupby_func)
```

```python
def line_is_correct(line):
    values_list = line.split(',')
    
    if len(values_list) != 3:
        return False
    
    column_1, column_2, column_3 = values_list
    
    if not str.isnumeric(column_2):
        return False
    
    if not column_3.strip():
        return False
    
    return True
```
