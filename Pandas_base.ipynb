```python
# Создание датафрейма
name = [choice(["Vlad", "Artem", "Ivan", "Sergey", "Arnold", "Valery"]) for i in range(100)]
sport = [choice(["Volleyball", "Football", "Tennis", "Crossfit", "Running"]) for i in range(100)]
hours = [choice([1, 2, 3, 4, 5]) for i in range(100)]
data = pd.DataFrame(
    {
        "name": name,
        "sport": sport,
        "hours": hours
    }
)
data.head(1)
```

```python
# Основные параметры о датафрейма
print("Тип:", type(data)); print("Количество строк:", len(data))
print(f"Размер: {data.shape[0]} строк, {data.shape[1]} столбцов")
```

```python
# Добавление нового столбца
data['minutes'] = data['hours'] * 60
data.head(1)
```

```python
# Обращение к конкретному столбцу (series)
print(data.minutes.head(1))
print(data['minutes'].head(1))
```

```python
# Основная статистика по датафрейму
data.describe()
```

```python
# Определение типа. Для столбца и всей таблицы могут применяться разные методы
type(data), type(data['name']), type(data.name.head())
```

```python
# Получение уникальных элементов столбца name
data['name'].unique()
```

```python
# Получение количества уникальных элементов столбца name
print("1-й способ:", len(data['name'].unique()))
print("2-й способ:", data['name'].nunique())
```

```python
# Распределение количества строк по значениям столбца sport
data['sport'].value_counts()
```

```python
# Распределение количества строк по значениям столбца sport (в процентах)
data['sport'].value_counts(normalize=True)
```

```python
# Выбор нескольких столбцов
data[['name', 'hours']].head(3)
```

```python
# Среднее количество часов занятий спортом
data['minutes'].mean()
```

```python
# Вывести строки, где количество минут выше среднего
data[data.minutes > data['minutes'].mean()].head()
```

```python
# Фильтр на несколько условий сразу
# | - условие ИЛИ
# & - and
# () | (() | () & ())
data[
    (data['name'] == 'Vlad') &
    (
        (data['sport'] == 'Volleyball') | (data['sport'] == 'Tennis')
    ) &
    (data['minutes'] >= data['minutes'].mean())]
```

```python
# Создание датафрейма с именами и возрастом
people = pd.DataFrame(
    {
        'name': ["Vlad", "Artem", "Ivan", "Sergey", "Arnold", "Valery"],
        'age': [22, 23, 26, 35, 49, 23]
    }
)
people
```

```python
people.columns
```

```python
data.columns
```

```python
# Объединение датафреймов. Ищет совпадения элемента правого фрейма с элементом в левом
pd.merge(people, data)
```

```python
# Что делать если названия столбцов не совпадают
people.columns = ['person_name', 'age']
print(people.columns); print(data.columns)
pd.merge(people, data, left_on='person_name', right_on='name').head()
```

```python
# Типы join: inner (default), outer, left, right
A = pd.DataFrame({'color': ['green', 'yellow', 'red'], 'num': [1, 2, 3]})
B = pd.DataFrame({'color': ['green', 'yellow', 'pink'], 'size': ['S', 'M', 'L']})
```

```python
A
```

```python
B
```

```python
# Только совпадающие элементы
pd.merge(A, B, how='inner')
```

```python
# Все элементы
pd.merge(A, B, how='outer')
```

```python
# Сохраняются все ключи из левого (A) фрейма, а из правого (B) совпадения
pd.merge(A, B, how='left')
```

```python
# Сохраняются все ключи из правого (B) фрейма, а из левого (А) совпадения
pd.merge(A, B, how='right')
```
